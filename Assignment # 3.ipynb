# PROGRAM FOR COUNTING WORDS, LINES, SPACES, TABS, PARAGRAPHS AND NON-SPACE CHARACTERS

```python
    """PROGRAM FOR COUNTING WORDS, LINES, SPACES, TABS, PARAGRAPHS AND NON-SPACE CHARACTERS"""

class file_parser:
   
        #initialiser of the class
        def __init__(self,file_name):
            self.words = 0
            self.spaces = 0
            self.lines = 0
            self.tabs = 0
            self.paragraphs = 0
            self.non_space_chars = 0
            self.filename = file_name
            


        #method for counting words
        def count_words(self): 
            contents = open(self.filename, encoding="utf8") 
            for line in contents:
                word = line.split()
                self.words += len(word)
            return "Number of words: " + str(self.words)


        #method for counting spaces
        def count_spaces(self): 
            contents = open(self.filename, encoding="utf8") 
            f = contents.read()
            for line in f: 
                if(line == " "):
                    self.spaces += 1
            return "Number of spaces: " + str(self.spaces)


        #method for counting lines
        def count_lines(self):
            contents = open(self.filename, encoding="utf8") 
            self.lines = len(contents.readlines())
            return "Number of lines: " + str(self.lines)


        #method for counting tabs
        def count_tabs(self):
            contents = open(self.filename, encoding="utf8") 
            f = contents.read()
            for line in f: 
                if(line == "\t"):
                    self.tabs += 1
            return "Number of tabs: " + str(self.tabs)


        #method for counting paragraphs
        def count_paragraphs(self):
            contents = open("sample_text_file.txt4", encoding="utf8") 
            content = contents.read().split("\n\t")
            self.paragraphs = len(content)
            return "Number of paragraphs: " + str(self.paragraphs)


        #method for counting non space characters
        def count_non_space_chars(self):
            contents = open(self.filename, encoding="utf8") 
            f = contents.read()  
            self.non_space_chars = len(f)-(self.tabs+self.spaces)
            return "Number of non space characters:" + str(self.non_space_chars)
        
        
        #method for priting all the details at once
        def details_of_the_file(self):
            print(" Details of the syntactic analysis are:")
            print("",count_words(self), "\n", count_spaces(self),"\n", count_lines(self), "\n", 
                  count_tabs(self),"\n",count_paragraphs(self),"\n", count_non_space_chars(self))
            
           
           

        

count = file_parser("sample_text_file.txt4")

count.details_of_the_file()
```

# PROGRAM FOR COUNTING AND NAMING THE HTML TAGS

```python
from html.parser import HTMLParser

#class for naming tags
class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        print("<" + tag + ">")

    def handle_endtag(self, tag):
        print("</" + tag + ">")


#class for counting tags
class counter:
    def __init__(self):
        self.tags = 0
    
    def count_tags(self):
        file = open("example_html_file.html") 
        contents = file.read()
        for tag in contents:
            if(tag == ">"):
                self.tags += 1
        return  "Number of tags: " + str(self.tags)
        


number_of_tags = counter()
print(number_of_tags.count_tags())
     


print("\nTags used are: ")
file = open("example_html_file.html")   
parser = MyHTMLParser()
parser.feed(file.read())
```
